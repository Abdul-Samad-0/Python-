```python

# Q5) Write a program to take a list and a number input from user and then delete all elements in the list less than that number.

num1 = int(input("please enter the number "))
new_list=[]
num=input("Enter integer list: ")
while num !=' ' :
   new_list.append(num)
   num=input("Enter integer list: ") 
print(new_list)
i: int = 0
while i < len(new_list) :
    if int(new_list[i]) <= num1 :
        del new_list[i]
    else:
        i=i+1
        
print(new_list)
```

```python
# Q7) Write a program that accepts a word from the user and reverses it using loop. For example, ‘Pakistan’ becomes‘natsikaP’.
word = input("Enter a word: ")
reversed_word = ""


for i in range(len(word)):
    reversed_word += word[len(word) - 1 - i]

print("The reversed word is: ", reversed_word)
```

```python
#Q8) Write a Python program that iterates the integers from 1 to 50. For multiples of three print "Fizz" instead of the number and for multiples of five print "Buzz". For numbers that are multiples of three and five, print "FizzBuzz".
   
for i in range(1,50):
   if i%3 and i%5 :
     print(" FizzBuzz ")
   elif i%3 :
     print(" Fizz ")
   elif i%5 :
     print(" BUZZ ")
```

```python
# Q10) Write a program to print a dictionary where the keys are numbers between 1 and 15 (both included) and the values are the square of the keys.

dictionary = {}
i = 1
while i <= 15:
    dictionary[i] = i ** 2
    i += 1

print(dictionary)
```
